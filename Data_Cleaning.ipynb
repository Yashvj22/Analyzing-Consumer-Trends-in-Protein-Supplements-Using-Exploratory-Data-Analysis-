## Scoops Filling NaN Values

```python
def fill_scoops_with_group_values(group):
    mode_values = group['scoops'].mode()
    if not mode_values.empty:
        common_value = mode_values.iloc[0] 
    else:
        common_value = group['scoops'].dropna().mean()  

    # Fill NaN values with the common value (only if common_value is valid)
    if pd.notna(common_value):
        group['scoops'] = group['scoops'].fillna(common_value)
    return group

# Apply the function to each group based on 'Net_Quantity'
df = df4.groupby('Net_Quantity', group_keys=False).apply(fill_scoops_with_group_values)
```

```python
## backward filling NaN values , beacuse there was same Brand which having NaN values in scoops
## This replacment didn't happen because these products are of single group , That's why in above step these NaN values doesn't replace

df["scoops"] = df["scoops"].bfill()
```

## Serving_Size filling NaN values

```python
df3[df3['Serving_Size'].isna()]
```

```python
def fill_serving_size(group):
    mode_values = group['Serving_Size'].mode()
    if not mode_values.empty:
        common_value = mode_values.iloc[0]  
    else:
        common_value = group['Serving_Size'].dropna().mean()  
    
    if pd.notna(common_value):
        group['Serving_Size'] = group['Serving_Size'].fillna(common_value)
    return group

# Apply the function to each group based on 'Net_Quantity' and 'Scoops'
df = df.groupby(['Net_Quantity', 'scoops'], group_keys=False).apply(fill_serving_size)
```

```python
## Now  only 58 rows are their, which having NaN values , from 220 we reduce to 58

df["Serving_Size"].isna().sum()
```

```python
## backward filling NaN values , beacuse there was same Brand which having NaN values in serving_size
## This replacment didn't happen because these products are of single group , That's why in above step these NaN values doesn't replace

df["Serving_Size"] = df["Serving_Size"].bfill()
```
